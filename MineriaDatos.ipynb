````python
# Importamos las bibliotecas
import numpy as np
import matplotlib.pyplot as plt

# Creamos datos de ejemplo
x = np.linspace(0, 10, 100)  # Crear un array de 100 puntos entre 0 y 10
y = np.sin(x)                # Calcular el seno de cada punto en x

# Crear un gráfico
plt.figure(figsize=(8, 6))   # Establecer el tamaño de la figura

# Agregar datos al gráfico
plt.plot(x, y, label='Seno(x)')  # Agregar la curva al gráfico y asignarle una etiqueta

# Agregar etiquetas y título
plt.xlabel('X')               # Etiqueta del eje X
plt.ylabel('Y')               # Etiqueta del eje Y
plt.title('Gráfico de Seno')  # Título del gráfico

# Agregar una leyenda
plt.legend()

# Mostrar el gráfico
plt.grid(True)  # Mostrar una cuadrícula en el gráfico
plt.show()      # Mostrar el gráfico en la salida del cuaderno
````
